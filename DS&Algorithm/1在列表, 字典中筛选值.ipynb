#### 在数组中筛选[0,12,-1,3,5,-19] 大于零的数

```python
l = [0,12,-1,3,5,-19]
```

##### for 循环写法

```python
l1 = []
for i in l:
    if i > 0:
        l1.append(i) 
print (l1)
```

```python
for index,item in enumerate(l):
    print("元素下标：",index,"元素值：",item)
```

##### 列表推导式

```python
l2 = [i for i in l if i > 0]
```

```python
l2
```

#### 效率上：for循环比列表推导式要慢，所以Python中推荐使用列表推导式
